```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split, RepeatedKFold, GridSearchCV
from sklearn.metrics import mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
```

```python
def season(date):
    month = date.month
    season = None
    if month in [3, 4, 5]:
        season = 'spring'
    elif month in [6, 7, 8]:
        season = 'summer'
    elif month in [9, 10, 11]:
        season = 'fall'
    elif month in [12, 1, 2]:
        season = 'winter'
    return season
```

# 2003 Daily data

## Pre-Processing

### Rule for merging hour -> daily
- max temp, 
- max snow depth (not reasonable)
- max snowfall
- max precip
- max irradiance
- max cloud coverage 
- max air density 
- sum all energy

```python
df_daily = pd.read_csv(os.getcwd() + '/Datasets/Ontario/Daily_Electricity_Ontario.csv')
```

```python
df_daily['time'] = pd.to_datetime(df_daily['time']).dt.date
df_daily['time'] = pd.to_datetime(df_daily['time'], format='%Y/%m/%d')
# Select the 'Total Energy Use from Electricity (MW)' column
energy_consumption_column = 'Total Energy Use from Electricity (MW)'

# Apply MinMaxScaler only to the energy consumption column
scaler = MinMaxScaler()
normalized_energy_consumption = scaler.fit_transform(df_daily[[energy_consumption_column]])

# Replace the original energy consumption column with the normalized one
df_daily[energy_consumption_column] = normalized_energy_consumption
```

```python
df_daily['season'] = df_daily.apply(lambda x: season(x['time']), axis=1)
df_daily = df_daily[['time', 'season', 'precipitation', 'temperature', 'irradiance_surface', 
           'irradiance_toa', 'snowfall', 'snow_depth', 'cloud_cover', 'air_density', 'Total Energy Use from Electricity (MW)']]
df_daily
```

```python
df_daily['season'].value_counts()
```

```python
df_daily['Total Energy Use from Electricity (MW)'].max()
```

```python
df_daily['Total Energy Use from Electricity (MW)'].mean()
```

```python
df_daily['Total Energy Use from Electricity (MW)'].min() ## too low
```

```python
df_daily.isnull().sum() # no Nan cell
```

## Predict 2016 energy consumption

```python
def get_period(df, start_date, end_date, col_name='time'):
    df_start = df[df[col_name] >= pd.to_datetime(start_date)]
    return df_start[df_start[col_name] <= pd.to_datetime(end_date)]
```

```python
start_date = '2016-01-01'
end_date = '2016-12-31'
df_2016 = get_period(df_daily, start_date, end_date)
df_2016
```

```python
x_train = np.array(df_daily[df_daily['time'] <= pd.Timestamp(2015, 12, 31)].drop(columns=["time", "season", "Total Energy Use from Electricity (MW)"]))
y_train = np.array(df_daily[df_daily['time'] <= pd.Timestamp(2015, 12, 31)]["Total Energy Use from Electricity (MW)"])

x_test = np.array(df_2016.drop(columns=["time", "season", "Total Energy Use from Electricity (MW)"]))
y_test = np.array(df_2016["Total Energy Use from Electricity (MW)"])
```

```python
print('Training Features Shape (X_train):', x_train.shape)
print('Training Labels Shape (y_train):', y_train.shape)
print('Testing Features Shape (X_test):', x_test.shape)
print('Testing Labels Shape (y_test):', y_test.shape)
```

```python
# decision tree regressor model
tree = DecisionTreeRegressor().fit(x_train, y_train)
# linear regression model
lr = LinearRegression().fit(x_train, y_train)
# random forest
rf = RandomForestRegressor().fit(x_train, y_train)
# gradient boosting
gb = GradientBoostingRegressor().fit(x_train, y_train)
```

```python
tree_pred = tree.predict(x_test)
lr_pred = lr.predict(x_test)
rf_pred = rf.predict(x_test)
gb_pred = gb.predict(x_test)
```

```python
def plot_scatter(y_test, y_pred, model_name):
    plt.figure(figsize=(8, 8))
    plt.scatter(y_test, y_pred, alpha=0.5)
    plt.xlabel("Actual Energy Use")
    plt.ylabel("Predicted Energy Use")
    plt.title(f"{model_name} - Predicted vs Actual Energy Use")
    # Plot a diagonal line for reference
    max_value = max(max(y_test), max(y_pred))
    plt.plot([0, max_value], [0, max_value], 'k--', lw=2)

    plt.show()
```

```python
models = ['Linear Regression', "Decision Tree Regressor", "Random Forest Regressor", "Gradient Boosting Regressor"]
y_preds = [lr_pred, tree_pred, rf_pred, gb_pred]

for i in range(len(models)):
    plot_scatter(y_test, y_preds[i], models[i])
```

```python
for i in range(len(models)):
    mae = mean_absolute_error(y_test, y_preds[i])
    print(f'MAE = {mae} for {models[i]} model')
```

- Best model so far - Gradient Boosting Regressor Model

# 2003 hourly data

## Pre-processing

```python
df_hourly = pd.read_csv(os.getcwd() + '/Datasets/Ontario/Hourly_Electricity_Ontario.csv')
```

```python
# Assuming 'time' and 'Total Energy Use from Electricity (MW)' columns are in the dataframe (df_hourly)
df_hourly['time'] = pd.to_datetime(df_hourly['time'])

# Select the 'Total Energy Use from Electricity (MW)' column
energy_consumption_column = 'Total Energy Use from Electricity (MW)'

# Apply MinMaxScaler only to the energy consumption column
scaler = MinMaxScaler()
normalized_energy_consumption = scaler.fit_transform(df_hourly[[energy_consumption_column]])

# Replace the original energy consumption column with the normalized one
df_hourly[energy_consumption_column] = normalized_energy_consumption

# Filter out the year 2003
df_hourly_2004 = df_hourly[df_hourly['time'].dt.year != 2003]

df_hourly
```

```python
# Separate the time column into dates and hour column 
df_hourly['dates'] = pd.to_datetime(df_hourly['time']).dt.date
df_hourly['hour'] = pd.to_datetime(df_hourly['time']).dt.hour

# Create the season column
df_hourly['season'] = df_hourly.apply(lambda x: season(x['dates']), axis=1)
```

### Add the season column 
- Spring: March 1st to May 31st
- Summer: June 1st to August 31th
- Fall: September 1st to November 30th
- Winter: December 1st to February 28/29th
- Based on https://www.durhamimmigration.ca/en/moving-to-durham-region/weather---four-seasons.aspx

```python
# Drop the time column and reorder the order of the columns
df_hourly.drop(columns='time')
df_hourly = df_hourly[['dates', 'hour', 'season', 'precipitation', 'temperature', 'irradiance_surface', 
           'irradiance_toa', 'snowfall', 'snow_depth', 'cloud_cover', 'air_density', 'Total Energy Use from Electricity (MW)']]
```

## four different models - ML Model

```python
input_data = np.array(df_hourly.drop(columns=["dates", "hour", "season", "Total Energy Use from Electricity (MW)"]))
output_data = np.array(df_hourly["Total Energy Use from Electricity (MW)"])
```

```python
# 0.1 -> 90% training data and 10% testing data
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size = 0.3)

# decision tree regressor model
tree = DecisionTreeRegressor().fit(x_train, y_train)
# linear regression model
lr = LinearRegression().fit(x_train, y_train)
# random forest
rf = RandomForestRegressor().fit(x_train, y_train)
# gradient boosting
gb = GradientBoostingRegressor().fit(x_train, y_train)
```

```python
tree_prediction = tree.predict(x_test)
lr_prediction = lr.predict(x_test)
rf_prediction = rf.predict(x_test)
gb_prediction = gb.predict(x_test)
```

```python
def plot_result(y_test, y_pred, df=None, model_type="Linear Regression", xlim_start=None, xlim_end=None):
    plt.figure(figsize=(12,7))
    plt.title(f'{model_type} Model')
    plt.xlabel('Days')
    plt.ylabel('Total Energy Use from Electricity (MW)')
    if df is not None:
        actual_line, = plt.plot(df['dates'], y_test, color="blue")
        predicted_line, = plt.plot(df['dates'], y_pred, color="red")
    else:
        actual_line, = plt.plot(y_test, color="blue")
        predicted_line, = plt.plot(y_pred, color="red")

    if xlim_start is not None and xlim_end is not None:
        plt.xlim(xlim_start, xlim_end)

        # Annotating the lines
    plt.annotate('Actual Energy Use', xy=(0.75, 0.85), xycoords='axes fraction', fontsize=12, color=actual_line.get_color())
    plt.annotate('Predicted Energy Use', xy=(0.75, 0.80), xycoords='axes fraction', fontsize=12, color=predicted_line.get_color())
    plt.show()
```

```python
plot_result(y_test=y_test, y_pred=lr_prediction, xlim_start=0, xlim_end=100)
```

```python
plot_result(y_test, tree_prediction, model_type="Decision Tree Regressor",xlim_start=0, xlim_end=100)
```

```python
plot_result(y_test, rf_prediction, model_type="Random Forest Regressor",xlim_start=0, xlim_end=100)
```

```python
plot_result(y_test, gb_prediction, model_type="Gradient Boosting Regressor",xlim_start=0, xlim_end=100)
```
